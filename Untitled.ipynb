```python
import numpy as np
import pandas as pd
from collections import defaultdict
from mitreattack.stix20 import MitreAttackData
import logging

# Configure logging
logging.basicConfig(level=logging.INFO)
logger = logging.getLogger(__name__)

# Load MITRE ATT&CK data
logger.info("Loading MITRE ATT&CK data...")
mitre_attack_data = MitreAttackData("enterprise-attack.json")

# Load the APT G0019 techniques data from CSV file
apt_techniques_df = pd.read_csv("./matrices/apt_g0019_techniques_data.csv")

# Define the canonical tactic order for MITRE ATT&CK Enterprise
canonical_tactic_order = [
    "Reconnaissance",
    "Resource Development",
    "Initial Access",
    "Execution",
    "Persistence",
    "Privilege Escalation",
    "Defense Evasion",
    "Credential Access",
    "Discovery",
    "Lateral Movement",
    "Collection",
    "Command and Control",
    "Exfiltration",
    "Impact"
]

def get_sorted_tactics(technique_id):
    """
    Retrieve and sort the tactics associated with a technique based on the canonical order.
    """
    tactics = mitre_attack_data.get_tactics_by_technique(technique_id)
    sorted_tactics = sorted(
        tactics,
        key=lambda t: canonical_tactic_order.index(t.name)
        if t.name in canonical_tactic_order else len(canonical_tactic_order)
    )
    return sorted_tactics

def get_primary_tactic_index(technique_id):
    """
    Return the index (in the canonical tactic order) of the technique's primary tactic.
    """
    sorted_tactics = get_sorted_tactics(technique_id)
    if sorted_tactics:
        return canonical_tactic_order.index(sorted_tactics[0].name)
    return None  # If no tactic found

def compute_adjusted_popularity(technique_ids, all_groups_usage, mitigations_mitigating):
    """
    Compute adjusted popularity for each technique based on group usage and mitigations.
    """
    popularity = {}
    mitigation_counts = {}
    
    # Calculate popularity (number of groups using each technique)
    for tech_id in technique_ids:
        pop = len(all_groups_usage.get(tech_id, []))
        popularity[tech_id] = pop if pop > 0 else 1  # default to 1 if not found
    
    # Calculate mitigation counts (number of mitigations for each technique)
    for tech_id in technique_ids:
        count = len(mitigations_mitigating.get(tech_id, []))
        mitigation_counts[tech_id] = count
    
    # Compute mitigation factor: techniques with more mitigations get a lower factor
    max_mitigation = max(mitigation_counts.values()) if mitigation_counts else 1
    mitigation_factor = {tech_id: 1 - (mitigation_counts[tech_id] / (max_mitigation + 1e-6))
                         for tech_id in technique_ids}
    
    # Compute adjusted popularity
    adjusted_popularity = {tech_id: popularity[tech_id] * mitigation_factor[tech_id]
                           for tech_id in technique_ids}
    
    # Normalize adjusted popularity
    max_adjusted_pop = max(adjusted_popularity.values())
    adjusted_popularity = {k: v / max_adjusted_pop for k, v in adjusted_popularity.items()}
    
    return adjusted_popularity

def build_transition_matrices(technique_ids, tactic_categories, platforms, adjusted_popularity):
    """
    Build success and failure transition matrices based on tactic categories and adjusted popularity.
    """
    n = len(technique_ids)
    failure_matrix = np.zeros((n, n))
    success_matrix = np.zeros((n, n))
    
    # Group techniques by their tactic category
    techniques_by_category = defaultdict(list)
    for idx, cat in enumerate(tactic_categories):
        techniques_by_category[cat].append(idx)
    
    # 1. Failure Transitions: next technique must be in the SAME tactic category and same platform
    for i in range(n):
        current_cat = tactic_categories[i]
        current_platforms = platforms[i]
        indices_in_cat = techniques_by_category[current_cat]
        valid_indices = [j for j in indices_in_cat if j != i and any(p in current_platforms for p in platforms[j])]
        if valid_indices:
            total_weight = sum(adjusted_popularity[technique_ids[j]] for j in valid_indices)
            if total_weight == 0:
                # Fall back to uniform distribution if all adjusted popularities are zero
                for j in valid_indices:
                    failure_matrix[i, j] = 1 / len(valid_indices)
            else:
                for j in valid_indices:
                    failure_matrix[i, j] = adjusted_popularity[technique_ids[j]] / total_weight
        else:
            failure_matrix[i, i] = 1.0
    
    # 2. Success Transitions: next technique must be in the NEXT available tactic category and same platform
    for i in range(n):
        current_cat = tactic_categories[i]
        current_cat_index = canonical_tactic_order.index(current_cat)
        next_category = None
        for cat in canonical_tactic_order[current_cat_index+1:]:
            if cat in techniques_by_category:
                next_category = cat
                break
        if next_category:
            indices_next = techniques_by_category[next_category]
            # Ensure the next tactic uses the same platform
            valid_indices_next = [j for j in indices_next if any(p in platforms[i] for p in platforms[j])]
            total_weight = sum(adjusted_popularity[technique_ids[j]] for j in valid_indices_next)
            if total_weight == 0:
                for j in valid_indices_next:
                    success_matrix[i, j] = 1 / len(valid_indices_next)
            else:
                for j in valid_indices_next:
                    success_matrix[i, j] = adjusted_popularity[technique_ids[j]] / total_weight
        else:
            success_matrix[i, i] = 1.0
    
    return success_matrix, failure_matrix

def main(group_stix_id):
    """
    Main function to compute transition matrices for a specific APT group.
    """
    logger.info(f"Processing APT group with STIX ID: {group_stix_id}")
    
    # Retrieve techniques used by the group
    techniques_used = mitre_attack_data.get_techniques_used_by_group(group_stix_id)
    technique_list = []
    for entry in techniques_used:
        tech = entry["object"]
        attack_id = mitre_attack_data.get_attack_id(tech.id)
        primary_index = get_primary_tactic_index(tech.id)
        if primary_index is None:
            continue  # Skip if no tactic available
        tactic_category = canonical_tactic_order[primary_index]
        technique_list.append((tech.id, attack_id, primary_index, tactic_category))
    
    # Sort techniques by their primary tactic index
    technique_list.sort(key=lambda x: x[2])
    technique_ids = [x[0] for x in technique_list]
    technique_names = [x[1] for x in technique_list]
    tactic_categories = [x[3] for x in technique_list]
    platforms = apt_techniques_df.loc[apt_techniques_df['id'].isin(technique_ids), 'platforms'].values
    n = len(technique_ids)
    
    # Compute adjusted popularity
    all_groups_usage = mitre_attack_data.get_all_groups_using_all_techniques()
    mitigations_mitigating = mitre_attack_data.get_all_mitigations_mitigating_all_techniques()
    adjusted_popularity = compute_adjusted_popularity(technique_ids, all_groups_usage, mitigations_mitigating)
    
    # Build transition matrices
    success_matrix, failure_matrix = build_transition_matrices(technique_ids, tactic_categories, platforms, adjusted_popularity)
    
    # Combine into a single matrix of tuples (success, failure)
    combined_matrix = np.empty((n, n), dtype=object)
    for i in range(n):
        for j in range(n):
            combined_matrix[i, j] = (success_matrix[i, j], failure_matrix[i, j])
    
    # Convert to DataFrame for visualization
    df_combined = pd.DataFrame(combined_matrix, index=technique_names, columns=technique_names)
    
    logger.info("Combined Transition Matrix (each cell is (success, failure)):")
    print(df_combined)

# Example usage
if __name__ == "__main__":
    group_stix_id = "intrusion-set--2a158b0a-7ef8-43cb-9985-bf34d1e12050"  # Example: G0019
    main(group_stix_id)
```

```python
def generate_attack_chains(success_matrix, failure_matrix, technique_names, tactic_categories, canonical_tactic_order):
    """
    Generate attack chains starting from each technique in the first tactic.
    At each step, it chooses the candidate with the highest probability.
    It first looks at success transitions (i.e. moving to the next tactic category).
    If none exist, it falls back to failure transitions (i.e. same tactic).
    
    Parameters:
        success_matrix (np.array): Matrix with success transition probabilities.
        failure_matrix (np.array): Matrix with failure transition probabilities.
        technique_names (list): List of technique names.
        tactic_categories (list): List of tactic categories corresponding to each technique.
        canonical_tactic_order (list): Predefined order of tactics.
    
    Returns:
        List of chains, where each chain is a list of indices corresponding to techniques.
    """
    # Identify starting techniques: those in the first tactic (e.g., "Reconnaissance")
    first_tactic = canonical_tactic_order[0]
    start_indices = [i for i, tactic in enumerate(tactic_categories) if tactic == first_tactic]
    
    chains = []
    
    def dfs(current_index, current_chain):
        # First, look for success candidates (next tactic category)
        success_candidates = [(j, success_matrix[current_index, j]) 
                              for j in range(len(technique_names)) 
                              if success_matrix[current_index, j] > 0]
        # Sort candidates by probability in descending order
        success_candidates.sort(key=lambda x: x[1], reverse=True)
        
        if success_candidates:
            # For each candidate (from highest to lowest)
            for j, prob in success_candidates:
                if j in current_chain:  # avoid cycles
                    continue
                current_chain.append(j)
                dfs(j, current_chain)
                current_chain.pop()  # backtrack after exploring the branch
            return  # We do not mix in failure candidates if success ones exist at this step
        else:
            # If no success candidates, try failure transitions (same tactic category)
            failure_candidates = [(j, failure_matrix[current_index, j])
                                  for j in range(len(technique_names))
                                  if failure_matrix[current_index, j] > 0]
            failure_candidates.sort(key=lambda x: x[1], reverse=True)
            
            if failure_candidates:
                for j, prob in failure_candidates:
                    if j in current_chain:
                        continue
                    current_chain.append(j)
                    dfs(j, current_chain)
                    current_chain.pop()
                return
            else:
                # No further candidates; record the current chain as complete.
                chains.append(current_chain.copy())
    
    # Start a chain for each technique in the first tactic.
    for start in start_indices:
        dfs(start, [start])
        
    return chains

# Example usage:
# Assuming success_matrix, failure_matrix, technique_names, tactic_categories,
# and canonical_tactic_order are defined and computed from your MITRE ATT&CK data.

chains_indices = generate_attack_chains(success_matrix, failure_matrix, technique_names, tactic_categories, canonical_tactic_order)
# Convert chains from indices to technique names for readability.
chains_names = [[technique_names[i] for i in chain] for chain in chains_indices]

for idx, chain in enumerate(chains_names, start=1):
    print(f"Attack Chain {idx}: {chain}")
```

```python
import pandas as pd

def generate_attack_chains(
    df: pd.DataFrame,
    current_technique: str,
    visited: set = None,
    chain_so_far: list = None,
    results: list = None,
    max_depth: int = None
):
    """
    Recursively generate all possible attack chains given a transition-probability matrix.
    
    :param df:               DataFrame where df.loc[row_technique, col_technique] = (p_success, p_failure).
    :param current_technique: Technique we are currently at.
    :param visited:          Set of techniques visited so far (to avoid cycles).
    :param chain_so_far:     The chain of techniques so far.
    :param results:          A list to store the final chains.
    :param max_depth:        (Optional) If you want to limit the depth of each chain to avoid huge expansions.
    :return:                 A list of all possible chains (each chain is a list of techniques).
    """
    if visited is None:
        visited = set()
    if chain_so_far is None:
        chain_so_far = []
    if results is None:
        results = []
        
    # Add the current technique to the chain
    chain_so_far.append(current_technique)
    visited.add(current_technique)
    
    # If we have a max_depth limit, stop if reached
    if max_depth is not None and len(chain_so_far) >= max_depth:
        results.append(chain_so_far.copy())
        return results

    # Separate out success and failure transitions
    row = df.loc[current_technique]
    success_candidates = []
    failure_candidates = []
    
    # Gather transitions from the row
    for next_technique in df.columns:
        p_success, p_failure = row[next_technique]
        # Only consider unvisited techniques to avoid cycles
        if next_technique not in visited:
            if p_success > 0:
                success_candidates.append((next_technique, p_success))
            if p_failure > 0:
                failure_candidates.append((next_technique, p_failure))

    # Sort by probability descending
    success_candidates.sort(key=lambda x: x[1], reverse=True)
    failure_candidates.sort(key=lambda x: x[1], reverse=True)

    # If there are no transitions to follow, record the chain as complete
    if not success_candidates and not failure_candidates:
        results.append(chain_so_far.copy())
    else:
        # Explore all success branches first (descending probability)
        for next_technique, prob_succ in success_candidates:
            generate_attack_chains(
                df,
                next_technique,
                visited.copy(),
                chain_so_far.copy(),
                results,
                max_depth
            )
        # Then explore all failure branches (descending probability)
        for next_technique, prob_fail in failure_candidates:
            generate_attack_chains(
                df,
                next_technique,
                visited.copy(),
                chain_so_far.copy(),
                results,
                max_depth
            )
    
    return results


# ---------------------------------------------
# Example usage:
# ---------------------------------------------

# Suppose df_combined is your DataFrame of shape (N_techniques, N_techniques),
# each cell = (p_success, p_failure).

# Let's say T1588.002 is an initial technique. Generate all chains starting there:
all_chains = generate_attack_chains(df_combined, "T1566.001")

print(f"Number of possible chains: {len(all_chains)}")
for chain in all_chains[:10]:  # Just show first 10
    print(" -> ".join(chain))


import pandas as pd

def generate_attack_chains(
    df: pd.DataFrame,
    current_technique: str,
    visited: set = None,
    chain_so_far: list = None,
    results: list = None,
    max_depth: int = None
):
    """
    Recursively generate all possible attack chains given a transition-probability matrix.
    
    :param df:               DataFrame where df.loc[row_technique, col_technique] = (p_success, p_failure).
    :param current_technique: Technique we are currently at.
    :param visited:          Set of techniques visited so far (to avoid cycles).
    :param chain_so_far:     The chain of techniques so far.
    :param results:          A list to store the final chains.
    :param max_depth:        (Optional) If you want to limit the depth of each chain to avoid huge expansions.
    :return:                 A list of all possible chains (each chain is a list of techniques).
    """
    if visited is None:
        visited = set()
    if chain_so_far is None:
        chain_so_far = []
    if results is None:
        results = []
        
    # Add the current technique to the chain
    chain_so_far.append(current_technique)
    visited.add(current_technique)
    
    # If we have a max_depth limit, stop if reached
    if max_depth is not None and len(chain_so_far) >= max_depth:
        results.append(chain_so_far.copy())
        return results

    # Separate out success and failure transitions
    row = df.loc[current_technique]
    success_candidates = []
    failure_candidates = []
    
    # Gather transitions from the row
    for next_technique in df.columns:
        p_success, p_failure = row[next_technique]
        # Only consider unvisited techniques to avoid cycles
        if next_technique not in visited:
            if p_success > 0:
                success_candidates.append((next_technique, p_success))
            if p_failure > 0:
                failure_candidates.append((next_technique, p_failure))

    # Sort by probability descending
    success_candidates.sort(key=lambda x: x[1], reverse=True)
    failure_candidates.sort(key=lambda x: x[1], reverse=True)

    # If there are no transitions to follow, record the chain as complete
    if not success_candidates and not failure_candidates:
        results.append(chain_so_far.copy())
    else:
        # Explore all success branches first (descending probability)
        for next_technique, prob_succ in success_candidates:
            generate_attack_chains(
                df,
                next_technique,
                visited.copy(),
                chain_so_far.copy(),
                results,
                max_depth
            )
        # Then explore all failure branches (descending probability)
        for next_technique, prob_fail in failure_candidates:
            generate_attack_chains(
                df,
                next_technique,
                visited.copy(),
                chain_so_far.copy(),
                results,
                max_depth
            )
    
    return results


# ---------------------------------------------
# Example usage:
# ---------------------------------------------

# Suppose df_combined is your DataFrame of shape (N_techniques, N_techniques),
# each cell = (p_success, p_failure).

# Let's say T1588.002 is an initial technique. Generate all chains starting there:
all_chains = generate_attack_chains(df_combined, "T1078.002")

print(f"Number of possible chains: {len(all_chains)}")
for chain in all_chains[:10]:  # Just show first 10
    print(" -> ".join(chain))
```
